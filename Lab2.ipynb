```python
num=[20,30,40]
it=iter(num)
print(next(it))
print(next(it))
print(next(it))
```

Creating list num.
List are iterable and iter() function converts the list into an iterator.
next() fetches the next value from the iterator.

```python
class MyNumbers:
    def __init__(self, limit):
        self.num = 0
        self.limit = limit

    def __iter__(self):
        return self

    def __next__(self):
        if self.num < self.limit:
            self.num += 1
            return self.num
        else:
            raise StopIteration


numbers = MyNumbers(5)
print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
```

Creating class MyNumbers.
init is a constructor that runs during class creation.
self.num starts at 0 and keeps track of the current number.
self.limit sets the limit or max value of the iterator.
iter makes the object iterable and returns self.
next returns the next number.
print(next(numbers)) moves the iterator forward.

```python
def square_generators(limit):
    for i in range(1,limit+1):
        yield i*1

gen = square_generators(4)

print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

square_generators is a generator function.. 
A function becomes a generator when it uses the keyword yield.
Unlike return, yield pauses the function and remembers its state.
yield sends a value one at a time.
print(next(gen)) move the function forward.

```python
def square_generators(limit):
    for i in range(1, limit + 1):
        yield i * i

for square in square_generators(4):
    print(square)
```

```python
def my_decorator(func):
    def wrapper():
        print("Function is about to run")
        func()
        print("Function has finished running")
    return wrapper
@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

my_decorator is a decorater function that takes another function as arugment.
In this case, func will be say_hello.
wrapper() is a nested function.
It adds extra behavior: code before func() runs and code after func() runs.
func() calls the original function (say_hello).
@my_decorator is syntactic sugar for  say_hello = my_decorator(say_hello).

```python
def my_decorator(func):
    def wrapper():
        print("Function is about to run")
        func()
        print("Function has finished running")
    return wrapper

def say_hello():
    print("Hello!")

say_hello=my_decorator(say_hello)
say_hello()
```

```python
Here say_hello=my_decorator(say_hello) is equivalent to using the @my_decorator syntax.
```
