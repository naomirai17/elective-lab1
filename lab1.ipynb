```python
def find_max(numbers):
    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num
nums = [2, 9, 0, 15, 40]
print("Maximum number:", find_max(nums))
```

```python
def area_of_circle(radius):
    pi = 3.14159
    area = pi * radius * radius
    return area
r = 5
print("Area of circle:", area_of_circle(r))
```
